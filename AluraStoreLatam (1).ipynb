### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
suma_Total_p1 = tienda["Precio"].sum()
suma_Total_p2 = tienda2["Precio"].sum()
suma_Total_p3 = tienda3["Precio"].sum()
suma_Total_p4 = tienda4["Precio"].sum()
```

```python
print(f"El valor total de la tienda 1 es de: {suma_Total_p1}")
print(f"El valor total de la tienda 2 es de: {suma_Total_p2}")
print(f"El valor total de la tienda 3 es de: {suma_Total_p3}")
print(f"El valor total de la tienda 4 es de: {suma_Total_p4}")
```

# 2. Ventas por categoría

```python
ventapc_tienda = tienda["Categoría del Producto"].value_counts()
ventapc_tienda2 = tienda2["Categoría del Producto"].value_counts()
ventapc_tienda3 = tienda3["Categoría del Producto"].value_counts()
ventapc_tienda4 = tienda4["Categoría del Producto"].value_counts()

print(f"Tienda 1, {ventapc_tienda}")
print(f"Tienda 2, {ventapc_tienda2}")
print(f"Tienda 3, {ventapc_tienda3}")
print(f"Tienda 4, {ventapc_tienda4}")
#tienda 1, categoria del producto.
```

```python
ventapc_tienda = tienda.groupby("Categoría del Producto")[["Precio"]].sum().sort_values("Precio", ascending=False)
ventapc_tienda2 = tienda2.groupby("Categoría del Producto")[["Precio"]].sum().sort_values("Precio", ascending=False)
ventapc_tienda3 = tienda3.groupby("Categoría del Producto")[["Precio"]].sum().sort_values("Precio", ascending=False)
ventapc_tienda4 = tienda4.groupby("Categoría del Producto")[["Precio"]].sum().sort_values("Precio", ascending=False)

print("Tienda 1:")
print(ventapc_tienda)
print("Tienda 2:")
print(ventapc_tienda2)
print("Tienda 3:")
print(ventapc_tienda3)
print("Tienda 4:")
print(ventapc_tienda4)
#Columna de la derecha = Precio
```

# 3. Calificación promedio de la tienda


```python
#Calculo de la calificacion promedio por cada tienda
prom_calificacion_tienda = tienda["Calificación"].mean()
prom_calificacion_tienda2 = tienda2["Calificación"].mean()
prom_calificacion_tienda3 = tienda3["Calificación"].mean()
prom_calificacion_tienda4 = tienda4["Calificación"].mean()

#los resultados serían
print(f"La calificacion promedio de la tienda 1 es de: {prom_calificacion_tienda: .3f}")
print(f"La calificacion promedio de la tienda 2 es de: {prom_calificacion_tienda2: .3f}")
print(f"La calificacion promedio de la tienda 3 es de: {prom_calificacion_tienda3: .3f}")
print(f"La calificacion promedio de la tienda 4 es de: {prom_calificacion_tienda4: .3f}")
```

# 4. Productos más y menos vendidos

```python
def productos_maxmin_vendidos(tiendas):
  #Agrupando por producto y sumar las facturaciones
  ventapp = tiendas.groupby("Producto")[["Precio"]].sum().sort_values("Precio", ascending=False)
  #Obteniendo producto con mayor facturación
  producto_mas_vendido = ventapp.idxmax(), ventapp.max()
  #Obteniendo producto con menor facturación
  producto_menos_vendido = ventapp.idxmin(), ventapp.min()
  return producto_mas_vendido, producto_menos_vendido
```

```python
#Calculo de productos mas y menos vendidos por cada tienda
max_min_tienda = productos_maxmin_vendidos(tienda)
max_min_tienda2 = productos_maxmin_vendidos(tienda2)
max_min_tienda3 = productos_maxmin_vendidos(tienda3)
max_min_tienda4 = productos_maxmin_vendidos(tienda4)

#Los resultados serían
print(f"Producto más vendido de Tienda 1: {max_min_tienda[0][0].item()} (Cantidad: {max_min_tienda[0][1].item()})")
print(f"Producto menos vendido de Tienda 1: {max_min_tienda[1][0].item()} (Cantidad: {max_min_tienda[1][1].item()})")

print(f"Producto más vendido de Tienda 2: {max_min_tienda2[0][0].item()} (Cantidad: {max_min_tienda2[0][1].item()})")
print(f"Producto menos vendido de Tienda 2: {max_min_tienda2[1][0].item()} (Cantidad: {max_min_tienda2[1][1].item()})")

print(f"Producto más vendido de Tienda 3: {max_min_tienda3[0][0].item()} (Cantidad: {max_min_tienda3[0][1].item()})")
print(f"Producto menos vendido de Tienda 3: {max_min_tienda3[1][0].item()} (Cantidad: {max_min_tienda3[1][1].item()})")

print(f"Producto más vendido de Tienda 4: {max_min_tienda4[0][0].item()} (Cantidad: {max_min_tienda4[0][1].item()})")
print(f"Producto menos vendido de Tienda 4: {max_min_tienda4[1][0].item()} (Cantidad: {max_min_tienda4[1][1].item()})")
```
